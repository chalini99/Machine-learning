```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
from PIL import Image

# Step 2 — Dataset Path
dataset_path = r"C:\Users\Chalini\Documents\ECG"  # use 'r' for raw path

train_dir = os.path.join(dataset_path, "train")
test_dir = os.path.join(dataset_path, "test")

print("Train folder:", train_dir)
print("Test folder:", test_dir)

# Step 3 — Data Preprocessing
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

train_data = train_datagen.flow_from_directory(
    directory=train_dir,
    target_size=(128,128),
    batch_size=32,
    class_mode='binary',
    subset='training'
)

val_data = train_datagen.flow_from_directory(
    directory=train_dir,
    target_size=(128,128),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)

# Step 4 — Build CNN Model
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# Step 5 — Train the Model
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=10
)

# Step 6 — Plot Accuracy and Loss
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.title("Accuracy Before & After Training")

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title("Loss Before & After Training")
plt.show()

# Step 7 — Confusion Matrix and Report
val_data.reset()
pred = model.predict(val_data)
predicted_classes = (pred > 0.5).astype("int32").flatten()

cm = confusion_matrix(val_data.classes, predicted_classes)
plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

print("Classification Report:")
print(classification_report(val_data.classes, predicted_classes))

# Step 8 — Upload and Predict New ECG Image
def predict_ecg_image(image_path):
    img = Image.open(image_path).resize((128,128))
    img = np.array(img)/255.0
    img = np.expand_dims(img, axis=0)
    prediction = model.predict(img)[0][0]

    if prediction > 0.5:
        print("High Risk of Heart Disease Detected")
    else:
        print("Normal / Low Risk ECG Signal")
    return prediction

# Example — Test with your ECG image
predict_ecg_image(r"C:\Users\Chalini\Documents\ECG\test\abnormal\961.jpg")

```

```python
import os, glob, random
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import itertools
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import backend as K
from tensorflow.keras.layers import (Input, Conv2D, MaxPooling2D, Flatten,
                                     Dense, Dropout, BatchNormalization, LeakyReLU, Activation)
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve, auc
import seaborn as sns

# ---------------------------
# 1. Configure your dataset path here (update to your local/colab path)
# ---------------------------
# Example: local Jupyter
dataset_path = r"C:\Users\Chalini\Documents\ECG"   # <- change if needed
train_dir = os.path.join(dataset_path, "train")
test_dir = os.path.join(dataset_path, "test")

print("Train dir exists:", os.path.exists(train_dir))
print("Test dir exists:", os.path.exists(test_dir))

# ---------------------------
# 2. Quick data-cleaning check: remove unreadable images
# ---------------------------
def remove_corrupted_images(folder):
    count = 0
    for ext in ("*.png","*.jpg","*.jpeg","*.bmp"):
        for p in glob.glob(os.path.join(folder, ext)):
            try:
                Image.open(p).verify()
            except Exception:
                try:
                    os.remove(p)
                    count += 1
                except Exception:
                    pass
    return count

# Run cleaning on each class folder (fast)
if os.path.exists(train_dir):
    for sub in os.listdir(train_dir):
        subp = os.path.join(train_dir, sub)
        if os.path.isdir(subp):
            removed = remove_corrupted_images(subp)
            if removed:
                print(f"Removed {removed} corrupt images from {subp}")

# ---------------------------
# 3. Detect classes and set multi/single/binary mode
# ---------------------------
classes = [d for d in os.listdir(train_dir) if os.path.isdir(os.path.join(train_dir, d))]
num_classes = len(classes)
print("Detected classes (train):", classes, "=> num_classes =", num_classes)

if num_classes == 0:
    raise ValueError("No class subfolders found in train directory. Please ensure train/<class_name>/images exist.")

is_binary = (num_classes == 2)

# ---------------------------
# 4. Data generators (no-augmentation vs augmentation) 
# We'll create two experiments: baseline (rescale only) and augmented
# ---------------------------
IMG_SIZE = (128,128)
BATCH = 32

no_aug_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)

aug_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=10,
    width_shift_range=0.07,
    height_shift_range=0.07,
    shear_range=0.05,
    zoom_range=0.05,
    horizontal_flip=False  # flipping signals may be wrong for ECG; keep False unless justified
)

def make_gens(datagen):
    if is_binary:
        class_mode = 'binary'
    else:
        class_mode = 'categorical'
    train_gen = datagen.flow_from_directory(
        train_dir,
        target_size=IMG_SIZE,
        batch_size=BATCH,
        class_mode=class_mode,
        subset='training',
        shuffle=True
    )
    val_gen = datagen.flow_from_directory(
        train_dir,
        target_size=IMG_SIZE,
        batch_size=BATCH,
        class_mode=class_mode,
        subset='validation',
        shuffle=False
    )
    return train_gen, val_gen

train_base, val_base = make_gens(no_aug_datagen)
train_aug, val_aug = make_gens(aug_datagen)

class_names = list(train_base.class_indices.keys())
print("Class indices:", train_base.class_indices)

# ---------------------------
# 5. Model factory with interchangeable activation
# ---------------------------
def build_cnn(input_shape=(128,128,3), activation='relu', num_classes=2, dropout=0.5):
    inp = Input(shape=input_shape)
    def act(x):
        if activation == 'leaky_relu':
            return LeakyReLU(alpha=0.1)(x)
        elif activation == 'swish':
            return Activation(tf.nn.swish)(x)
        else:
            return Activation(activation)(x)
    x = Conv2D(32, (3,3), padding='same')(inp); x = act(x); x = BatchNormalization()(x); x = MaxPooling2D()(x)
    x = Conv2D(64, (3,3), padding='same')(x); x = act(x); x = BatchNormalization()(x); x = MaxPooling2D()(x)
    x = Conv2D(128, (3,3), padding='same')(x); x = act(x); x = BatchNormalization()(x); x = MaxPooling2D()(x)
    x = Flatten()(x)
    x = Dense(256)(x); x = act(x); x = Dropout(dropout)(x)
    if num_classes == 2:
        out = Dense(1, activation='sigmoid')(x)
        model = Model(inp, out)
        model.compile(optimizer=Adam(1e-4), loss='binary_crossentropy', metrics=['accuracy'])
    else:
        out = Dense(num_classes, activation='softmax')(x)
        model = Model(inp, out)
        model.compile(optimizer=Adam(1e-4), loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# ---------------------------
# 6. Training helper (with optional steps_per_epoch to keep runs short)
# ---------------------------
def train_model(model, train_gen, val_gen, epochs=6, steps_per_epoch=None, validation_steps=None):
    if steps_per_epoch is None:
        steps_per_epoch = min(train_gen.n // train_gen.batch_size, 200)  # cap for speed
    if validation_steps is None:
        validation_steps = min(val_gen.n // val_gen.batch_size, 50)
    history = model.fit(
        train_gen,
        steps_per_epoch=steps_per_epoch,
        validation_data=val_gen,
        validation_steps=validation_steps,
        epochs=epochs
    )
    return history

# ---------------------------
# 7. Baseline vs Augmented comparison (single activation e.g. relu)
# ---------------------------
print("\n--- Running baseline (no augmentation) with RELU ---")
model_base = build_cnn(activation='relu', num_classes=(2 if is_binary else num_classes))
hist_base = train_model(model_base, train_base, val_base, epochs=4)

print("\n--- Running augmented training with RELU ---")
model_aug = build_cnn(activation='relu', num_classes=(2 if is_binary else num_classes))
hist_aug = train_model(model_aug, train_aug, val_aug, epochs=4)

# Plot before/after (baseline vs augmented)
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(hist_base.history['val_accuracy'], label='baseline val_acc')
plt.plot(hist_aug.history['val_accuracy'], label='aug val_acc')
plt.legend(); plt.title("Validation Accuracy: baseline vs augmented")

plt.subplot(1,2,2)
plt.plot(hist_base.history['val_loss'], label='baseline val_loss')
plt.plot(hist_aug.history['val_loss'], label='aug val_loss')
plt.legend(); plt.title("Validation Loss: baseline vs augmented")
plt.show()

# ---------------------------
# 8. Activation-function comparison (small runs for each activation)
# ---------------------------
activations = ['relu', 'leaky_relu', 'elu', 'swish']
results = []

for act in activations:
    print(f"\n>>> Training with activation = {act}")
    m = build_cnn(activation=act, num_classes=(2 if is_binary else num_classes))
    h = train_model(m, train_aug, val_aug, epochs=3)  # keep small for quick comparison
    # evaluate on full validation set (predict)
    val_steps = val_aug.n // val_aug.batch_size
    if val_steps == 0:
        val_steps = 1
    val_aug.reset()
    preds = m.predict(val_aug, steps=val_steps, verbose=0)
    # get labels and compute metrics
    if is_binary:
        y_true = val_aug.classes[:len(preds)]
        y_prob = preds.ravel()[:len(y_true)]
        y_pred = (y_prob >= 0.5).astype(int)
        auc_score = roc_auc_score(y_true, y_prob) if len(np.unique(y_true))>1 else float('nan')
    else:
        y_true = val_aug.classes[:len(preds)]
        # preds is (N, num_classes)
        y_prob = preds[:len(y_true)]
        y_pred = np.argmax(y_prob, axis=1)
        try:
            y_true_oh = tf.keras.utils.to_categorical(y_true, num_classes=num_classes)
            auc_score = roc_auc_score(y_true_oh, y_prob, multi_class='ovr', average='macro')
        except Exception:
            auc_score = float('nan')
    acc = np.mean(y_pred == y_true)
    results.append({'activation':act,'accuracy':acc,'auc':auc_score})

# show results table and bar chart
import pandas as pd
df = pd.DataFrame(results)
print(df)
plt.figure(figsize=(6,4))
plt.bar(df['activation'], df['accuracy'])
plt.title("Activation comparison - Validation accuracy")
plt.ylabel("accuracy")
plt.show()

# ---------------------------
# 9. Detailed evaluation (confusion matrix + classification report) for the best model
# Choose the model with highest accuracy from the results (fallback to model_aug)
# ---------------------------
best_idx = int(np.argmax([r['accuracy'] for r in results])) if results else 0
best_activation = results[best_idx]['activation'] if results else 'relu'
print("Best activation (quick test):", best_activation)

best_model = build_cnn(activation=best_activation, num_classes=(2 if is_binary else num_classes))
_ = train_model(best_model, train_aug, val_aug, epochs=4)

# Predict on validation (full)
val_aug.reset()
val_steps = val_aug.n // val_aug.batch_size
preds_full = best_model.predict(val_aug, steps=val_steps+1, verbose=0)
if is_binary:
    y_prob = preds_full.ravel()[:val_aug.n]
    y_pred = (y_prob >= 0.5).astype(int)
else:
    y_prob = preds_full[:val_aug.n]
    y_pred = np.argmax(y_prob, axis=1)

y_true = val_aug.classes[:len(y_pred)]
print("\nClassification report:")
print(classification_report(y_true, y_pred, target_names=class_names))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names, cmap='Blues')
plt.xlabel("Predicted"); plt.ylabel("True"); plt.title("Confusion Matrix (Validation)")
plt.show()

# Plot ROC (binary) or per-class ROC (multi) if available
if is_binary and len(np.unique(y_true))>1:
    fpr, tpr, _ = roc_curve(y_true, y_prob)
    plt.figure(); plt.plot(fpr,tpr); plt.plot([0,1],[0,1],'--'); plt.title("ROC Curve (Validation)"); plt.show()
elif not is_binary:
    # compute macro ROC AUC per class
    try:
        y_true_oh = tf.keras.utils.to_categorical(y_true, num_classes=num_classes)
        for i, cname in enumerate(class_names):
            fpr, tpr, _ = roc_curve(y_true_oh[:,i], y_prob[:,i])
            plt.plot(fpr,tpr,label=cname)
        plt.plot([0,1],[0,1],'--'); plt.legend(); plt.title("Per-class ROC (Validation)"); plt.show()
    except Exception:
        pass

# ---------------------------
# 10. Grad-CAM interpretability (visual backtracking)
# ---------------------------
def find_last_conv_layer(model):
    for layer in reversed(model.layers):
        if isinstance(layer, Conv2D):
            return layer.name
    raise ValueError("No Conv2D layer found")

def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    grad_model = Model([model.inputs], [model.get_layer(last_conv_layer_name).output, model.output])
    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(predictions[0])
        loss = predictions[:, pred_index]
    grads = tape.gradient(loss, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0,1,2))
    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)
    heatmap = tf.maximum(heatmap, 0) / (tf.math.reduce_max(heatmap) + 1e-8)
    return heatmap.numpy()

def display_gradcam_on_image(img_path, heatmap, alpha=0.4):
    img = Image.open(img_path).convert("RGB").resize(IMG_SIZE)
    img_arr = np.array(img)/255.0
    plt.figure(figsize=(6,4))
    plt.imshow(img_arr)
    plt.imshow(heatmap, cmap='jet', alpha=alpha, extent=(0,IMG_SIZE[0],IMG_SIZE[1],0))
    plt.axis('off')
    plt.show()

# example: pick a validation file and show Grad-CAM
# get one sample path from val_aug
val_aug.reset()
sample_index = 0
if val_aug.n > 0:
    sample_batch, sample_labels = next(iter(val_aug))
    img_array = sample_batch[0:1]
    last_conv = find_last_conv_layer(best_model)
    heatmap = make_gradcam_heatmap(img_array, best_model, last_conv)
    # display overlay
    display_gradcam_on_image(val_aug.filepaths[0], heatmap)

# ---------------------------
# 11. PIL single-image prediction helper (with optional Grad-CAM overlay)
# ---------------------------
def predict_ecg_image(image_path, model=best_model, show_gradcam=True):
    img = Image.open(image_path).convert("RGB").resize(IMG_SIZE)
    arr = np.array(img)/255.0
    x = np.expand_dims(arr, axis=0)
    pred = model.predict(x)[0]
    if is_binary:
        prob = float(pred[0]) if pred.shape else float(pred)
        label = class_names[int(prob >= 0.5)]
        print(f"Predicted: {label} (prob={prob:.3f})")
    else:
        prob = pred
        label = class_names[np.argmax(prob)]
        print(f"Predicted: {label} (probs={np.round(prob,3)})")
    if show_gradcam:
        last_conv = find_last_conv_layer(model)
        hm = make_gradcam_heatmap(x, model, last_conv)
        display_gradcam_on_image(image_path, hm)
    return label

# Example usage:
# predict_ecg_image(r"C:\Users\Chalini\Documents\ECG\test\96.jpg")

# ---------------------------
# 12. Save best model and histories (optional)
# ---------------------------
# best_model.save("ecg_cnn_best.h5")
# Done.
```
